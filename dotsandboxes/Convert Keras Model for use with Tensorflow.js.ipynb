# **Convert Keras model for use with TensorFlow.js**

```
import os
import tensorflowjs as tfjs
from dotsandboxes.keras.NNet import NNetWrapper as nn
from dotsandboxes.DotsAndBoxesGame import DotsAndBoxesGame
```

```
game = DotsAndBoxesGame(n=3)
nnet = nn(game)
nnet.load_checkpoint(os.path.join('pretrained_models', 'dotsandboxes', 'keras', '3x3'), 'best.pth.tar')
```

```
tfjs.converters.save_keras_model(nnet.nnet.model, '/content/tfjs/')
```
